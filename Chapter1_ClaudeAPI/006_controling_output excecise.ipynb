````python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
````

### ヘルパー関数 : stop_sequence対応

````python
def add_user_message(message, text):
    user_message = {
        "role": "user",
        "content": text
    }
    message.append(user_message)

def add_assistant_message(message, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    message.append(assistant_message)

# Make a request
def chat(messages, system=None,temperature=1.0,stop_sequences=[]):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }
    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
````

### 出力制御の演習
- prefillingとストップシーケンスを利用して３つのボディだけを１回のレスポンスで得る
- ３つはコメントや説明を含んではいけない
- ヒント：prefillingは ``` のような文字列に限定されない

````python
messages = []

prompt = """
Generate three different sample AWS CLI commands. Each should be very short.
"""
add_user_message(messages, prompt)
print(messages[-1]['content'])

prefill = "Here are all three commands in a single block without any comments.``bash"
add_assistant_message(messages, prefill)

answer = chat(messages, stop_sequences=["```"])
print(answer)
````
